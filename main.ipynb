## 2. Now write a function that returns the largest element on a list

```python
#your code here
def greatest(lst):
    g = lst[0]
    for n in lst:
        if n > g:
            g = n
    
    return g
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n in [0,1]:
        return 1
    else:
        res = n * factorial(n-1)
        return res
#your code here
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
    
    # create dictionary of counters for each element in list
    counters = {}
    for e in arr:
        if not counters.get(e):
            counters[e] = 1
        else:
            counters[e] +=1
    
    # get list of counter values
    freqlist = list(counters.values())
    # get list of keys
    keylist = list(counters.keys())
    # identify index of largest value
    largest_idx = freqlist.index(greatest(freqlist))
    # return the key at the index for largest value
    return keylist[largest_idx]
    
mode_counter(['a','gr',2,3,4,3,'gr','gr'])
    
#your code here
```

```python
# population standard deviation
# sqrt(sum((x-mean)**2)/number of observations)

# sample standard deviation
# sqrt(sum((x-mean)**2)/number of observations - 1)


def st_dev(list_sd):
    # get N
    l_of_list = length(list_sd)
    # get mean of distribution
    mean = sum_all(list_sd)/l_of_list
    # return square root of ((sum of squared differences between each value in distribution from mean) divided by N)
    top = sum_all([(x-mean)**2 for x in list_sd])
    variance = top/(l_of_list-1)
    return variance**(1/2)
        

tlist=[1,2]
print(st_dev(tlist))
print(stdev(tlist))
#your code here

```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    # build a list of all ascii letters (only lower case for now?)
    # to save some typing work, I create a range and then convert that to strings
    alphabet = [chr(c) for c in range(97,123)]
    for l in alphabet:
        if l not in string.lower():
            return False
    
    return True
#your code here

pangram('The quick brown fox jumps over the lazy dog.')
pangram('test')
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python

def sort_alpha(string):
    tmpstring=''
    lst=[]
    for c in string:
        if c ==' ':
            continue
        elif c != ',':
            tmpstring += c
        else:
            lst.append(tmpstring)
            tmpstring = ''
        
    lst.append(tmpstring)
    lst = sorted(lst)
    
    targetstring = lst[0]
    for s in lst[1:]:
        # nicer option with spaces commented out to accomodate test algorithm
 #       targetstring += ', '
        targetstring += ','
        targetstring += s
        
    
    return(targetstring)
#your code here
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    low=False
    up=False
    num=False
    spec=False
    min8=False
    
    counter = 0
    special_chars = '!¬ß^$%&@/()=?+"\[]}{*#-_.,;:<>|'
        
    for c in password:
        counter+=1
        if counter > 7:
            min8 = True
        if ord(c) >=64 and ord(c)<=90:
            up=True
        elif ord(c) >=97 and ord(c)<=122:
            low=True
        elif ord(c) >=48 and ord(c)<=57:
            num=True
        elif c in special_chars:        # ok, I'm not going to test all special char ASCII areas separately
            spec=True
        
        if low==up==num==spec==min8==True:
            return True
    
    return False
            
#your code here

```
